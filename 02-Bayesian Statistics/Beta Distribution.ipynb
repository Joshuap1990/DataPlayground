# What is the beta distribution? 

The beta distribution is a continuous probability distribution.

The beta distribution can be understood as representing the probability distribution of probabilities - that is, represents all the possible values of a probability when we don't know what that probability is.

#### Why might we want to use the Beta distribution?

This distribution is only defined in Theta lying in the range [0,1] Because of this a beta distribution is a very natural distribution to use when we are talking about probabilities

#### Example use of a Beta Distribution
Imagine the case where we are flipping a coin, and we want to know how biased it is (if it is biased at all). If we flipped the coin 4 times, and got 3 heads - does that mean that the coin is definitely biased with a Pheads of 0.75 exactly? Of course not!

The way we represent that level of bias in the coin is the beta distribution. The beta distribution is defined by two coefficients a and b.

 - a is the number of times the event (coin landing on heads) occurred.

 - b is the number of times the event did not occur.




\begin{align*}
PDF = (x^(a-1)*(1-x)^(b-1))/Beta(a,b)
\end{align*}

### Viewing the beta distribution
Lets have a look at the beta distribution of that coin flipping example. As we have just discussed we observed 3 heads (a=3) and 1 tails (b=1). We'll be using the distributions in scipy.stats to plot the distributions

```python
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np
```

Define the beta distribution by its two coefficients

```python
a, b = 3,1
label = 'a = 3, b = 1'
def plot_beta(a,b,label):
    beta_distribution = beta(a, b)
    x_vals = np.linspace(0, 1, num=100)
    p_vals = beta_distribution.pdf(x_vals)
    plt.plot(x_vals, p_vals, label = label)
    plt.legend()
    plt.title('Beta Distribution')
    plt.xlabel('Theta')
    plt.ylabel('P')
    plt.xlim([0,1])
plot_beta(a,b,label)
```

#### Having more confidence in results with more coin flips

So far, the distribution above (Beta(a,b)) doesn't really tell us that much. the confidence interval is so wide, it looks like the most likely value is 1 (even though the inital data seemed to indicate 0.75).

Lets pretend for a moment that we have stared into a crystal ball and know the true value is 0.75. Lets keep on flipping coins (collecting data)

```python
new_samples = [[3,1,'a=3 b=1'],
              [6,2,'a=6 b=2'],
              [9,3,'a=9 b=3'],
              [18,6,'a=18 b=6'],]

for sample in new_samples:
    plot_beta(sample[0], sample[1], sample[2])
```

So what do we notice? We can see that as we flip more coins, the distribution becomes narrower and taller. We are becoming more and more confident in a particular set of values.

Remember we don't know the true answer from the crystal ball, all we can say is that the true answer lies within a range of values.

## Exploring the Beta Distribution

Next we will have some sliders to change the 
a and b parameters in the beta distribution and observe how that affects the shape of the beta distribution

```python
from ipywidgets import interact
import ipywidgets as widgets
```

```python

def beta_explore(a,b):
    #b=10
    from scipy.stats import beta
    import numpy as np
    import matplotlib.pyplot as plt
    
    beta_distribution = beta(a, b)

    x_vals = np.linspace(0, 1, num=40)
    p_vals = beta_distribution.pdf(x_vals)
    
    plt.plot(x_vals, p_vals)
    plt.xlim([0,1])
    plt.ylim([0,5])
    plt.title('Exploring the Beta Distribution')
    plt.show()
```

```python
interact(beta_explore, a=5,b=5)
```
